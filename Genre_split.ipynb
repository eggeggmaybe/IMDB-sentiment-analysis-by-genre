```python
def pmi(word1, word2, genre):
    pmi_path = r'C:/Users/Administrator/Desktop/MBS/Module3/Text and Web Analytics/assignment/My work/sort_reviews_by_genres/genres/train/for_pmi_only/'
    index = idx2word_inverse[word]
    word1_count = 0
    word2_count = 0
    together_count = 0
    total_word = 0      
    review_files = os.listdir(os.path.join(pmi_path, genre))
    for i in range(len(review_files)):
        temp_dict = Counter()
        with open(os.path.join(pmi_path, genre)+ '/' + review_files[i], 'r', encoding='utf-8') as f:
            for line in f:
                review = re.sub(r'[^\w\s]','',line)
            review_list = re.split(r'\s', review)
            total_word += len(review_list)
            word1_temp = 0
            word2_temp = 0
            for i in review_list:
                if i == word1:
                    word1_temp += 1
                    word1_count += 1
                if i == word2:
                    word2_temp += 1
                    word2_count += 1
            together_count += min(word1_temp, word2_temp)
    if word1_count == 0:
        output = f'{word1} is not in the corpus'
    elif word2_count == 0:
        output = f'{word2} is not in the corpus'
    elif together_count == 0:
        output = f"'{word1}' and '{word2}' never appear together in the {genre}"
    else:
        output = math.log2(total_word * together_count / word1_count / word2_count)

    return output
```

```python
def SO(word_i, positive_word, negative_word, genre):
    try:
        output = (pmi(word_i, positive_word, genre) - pmi(word_i, negative_word, genre))
        return output
    except:
        print(f'one of the word count is 0')
```

```python
def lex_update(SO_dict, genre):
    new_lexicon_path = 'C:/Users/Administrator/Desktop/MBS/Module3/Text and Web Analytics/assignment/My work/sort_reviews_by_genres/genres/train/lexicon/'
    with open(os.path.join(new_lexicon_path, genre) + '.txt', 'w', encoding='utf-8') as f:
        for word, rating in word_rating.items():
            if word not in SO_dict:
                f.write(str(rating) + '\n')
            else:
                if (float(rating) * float(SO_dict[word])) < 0:
                    rating = -1 * float(rating)
                    f.write(str(rating) + '\n')
                else:
                    f.write(str(rating) + '\n')
    return None
```
