```python
# imdb.vocab: reference for vocabs order (e.g. 'the' is order 0 and 'of' is order 3) 
# imdbEr.txt: expected rating for token in imdb.vocab (same order)
# labeledBow: bag of words for pos then neg reviews in training/test folder (ordered sequential, first pos then neg)
# reviews for any movie will not exceed 30 for uncorrelated rating purpose
# score <= 4 is classified as neg and score >= 7 is pos
# neutral ratings (5 to 6 inclusive) is excluded

# train/unsup has even number of reviews rating >5 and <=5 (note that ratings in this folder are all muted to 0)

# url files contains url which has the same order as in the rating folders
```

```python
from urllib.request import urlopen
import re
from bs4 import BeautifulSoup, SoupStrainer
import json
from collections import defaultdict
import codecs
#import sys
#import io
```

```python
def url_lookup(text_file):
    url_list = []
    with open(text_file, 'r') as f:
        for url in f:
            url.strip()
            url_temp = re.findall(r'(http:.+)usercomments', url)
            url_list.append(url_temp[0])
    return url_list

train_neg = url_lookup('train_urls_neg.txt')
train_pos = url_lookup('train_urls_pos.txt')
test_neg = url_lookup('test_urls_neg.txt')
test_pos = url_lookup('test_urls_pos.txt')
train_unsup = url_lookup('train_urls_unsup.txt')
```

```python
%%time
# if error occurs, use 'continue version' to continue the job (BUT NEED TO FIND THE RIGHT NUMBER OF URL TO CONTINUE!)
# function input explain: 
# url_list is a .txt file containing the URLs
# title_output_name is to define the output .txt file name of titles
# genre_output_name is to define the output .txt file name of genres

def movie_info(url_list, title_output_name, genre_output_name):
    # the function is returning a list
    # index = 0 refers to title
    # index = 1 refers to genre
    
    #sys.stdout = io.TextIOWrapper(sys.stdout.buffer.encoding = 'utf-8') # DON'T ENABLE THIS LINE, made for future testing
    title_all = defaultdict(lambda: 'My default')
    genre_all = defaultdict(list)
    n = 1
    f_title = codecs.open(title_output_name, 'w', 'utf-8')
    f_title.close()
    f_genre = codecs.open(genre_output_name, 'w', 'utf-8')
    f_genre.close()
    current_url = ''
#    attr_dict = {'type': 'application/ld+json'} ### Soupstrainer ###
#    strainer = SoupStrainer('script', attr_dict) ### Soupstrainer ###
    
    for url in url_list:
#        soup = BeautifulSoup(urlopen(url), "lxml", parse_only = strainer) ### Soupstrainer ###
#        title = BeautifulSoup(urlopen(url), "lxml").title.string[:-7] # including the year but excluding '- imdb' ### Soupstrainer ###
        if url == current_url:
            f_title = codecs.open(title_output_name, 'a', 'utf-8')
            f_title.write(f'{title}\n')
            f_title.close()
            title_all[n] = title # append to dictionary where key is the order
            f_genre = codecs.open(genre_output_name, 'a', 'utf-8')
            f_genre.write(f'{genre_raw}\n')
            f_genre.close()
            genre_all[n] = genre # append to dictionary where key is the order
            current_url = url
            n += 1
        else:
            # use try and except beacuse some pages may no longer accessable
            # if this is the case, then 'unknown' will be assigned to both title and genre
            try:
                soup = BeautifulSoup(urlopen(url), "lxml")
                title = soup.title.string[:-7] # including the year but excluding '- imdb'
                soup_genre = soup.find_all('script', attrs = {'type': 'application/ld+json'})
                genre_raw = re.findall(r'"genre":(.*?\])', str(soup_genre[0])) # this is a string type extracted from html
                if len(genre_raw) == 0:
                    genre_raw = '["Unknown"]' # 'unknown' is the ones don't have genres on IMDb website
                else:
                    genre_raw = genre_raw[0]
                genre = json.loads(genre_raw) # this is a list containing the genres
                f_title = codecs.open(title_output_name, 'a', 'utf-8')
                f_title.write(f'{title}\n')
                f_title.close()
                title_all[n] = title # append to dictionary where key is the order
                f_genre = codecs.open(genre_output_name, 'a', 'utf-8')
                f_genre.write(f'{genre_raw}\n')
                f_genre.close()
                genre_all[n] = genre # append to dictionary where key is the order     
                current_url = url
                n += 1
            except:
                f_title = codecs.open(title_output_name, 'a', 'utf-8')
                f_title.write('Unknown\n')
                f_title.close()
                title = 'Unknown'
                title_all[n] = title # append to dictionary where key is the order
                f_genre = codecs.open(genre_output_name, 'a', 'utf-8')
                f_genre.write('["Unknown"]\n')
                f_genre.close()
                genre_raw = '["Unknown"]'
                genre = ['Unknown']
                genre_all[n] = genre # append to dictionary where key is the order     
                current_url = url
                n += 1

    return [title_all, genre_all]

#movie_info(train_neg, 'train_title_neg.txt', 'train_genre_neg.txt')
#movie_info(train_pos, 'train_title_pos.txt', 'train_genre_pos.txt')
#movie_info(test_neg, 'test_title_neg.txt', 'test_genre_neg.txt')
#movie_info(test_pos, 'test_title_pos.txt', 'test_genre_pos.txt')
movie_info(train_unsup, 'train_title_unsup.txt', 'train_genre_unsup.txt')
```
